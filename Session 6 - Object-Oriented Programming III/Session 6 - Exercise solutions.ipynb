## Exericse 1

```python
class Accumulator():

    def __init__(self, start=0):
        self._sum = start

    def __call__(self, x):
        self._sum += x
        print(f"I've now added {x} to my sum")

    @property
    def current_sum(self):
        return self._sum
```

```python
# initializing two instances with different starting values
adder_A = Accumulator(start=2)
adder_B = Accumulator(start=3)
```

```python
adder_A(3)
```

```python
adder_B(4)
```

```python
adder_A(5)
```

```python
adder_B(6)
```

```python
adder_A.current_sum
```

```python
adder_B.current_sum
```

## Exericse 2

```python
class Count_to_10():
    
    def __init__(self, start=0, step=1):
        self.num = start
        self.step = step
        
    def __iter__(self):
        return self
    
    def __next__(self):
        self.num += self.step
        if self.num > 10: 
            print('I can only count to 10! :(')
            raise(StopIteration)
        return self.num
    
    def set_step(self, step):
        self.step = step
```

```python
counter = Count_to_10(start=0, step=1)
```

```python
next(counter)
```

```python
next(counter)
```

```python
counter.set_step(2)
```

```python
for count in counter:
    print(count)
```
