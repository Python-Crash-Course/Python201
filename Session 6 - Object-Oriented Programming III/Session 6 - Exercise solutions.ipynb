```python
class Count_to_10():
    
    def __init__(self, start=0, step=1):
        self.num = start
        self.step = step
        
    def __iter__(self):
        return self
    
    def __next__(self):
        self.num += self.step
        if self.num > 10: 
            print('I can only count to 10! :(')
            raise(StopIteration)
        return self.num
    
    def set_step(self, step):
        self.step = step
```

```python
counter = Count_to_10(start=0, step=1)
```

```python
next(counter)
```

```python
next(counter)
```

```python
counter.set_step(2)
```

```python
for count in counter:
    print(count)
```
